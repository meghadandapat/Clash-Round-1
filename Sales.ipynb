```python
import pandas as pd
```

```python
df = pd.read_csv('./company_sales_data.csv')
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df['month_number'], df['total_units'], color='red', lineStyle='dashed', linewidth=3, marker='o', markerfacecolor='black')
plt.legend(["profit data of last year"], loc ="lower right")
plt.xlabel('Month Number')
plt.ylabel('No of units sold')
plt.xlabel('Company Sales Data of last year')
```

```python
plt.plot(df['month_number'], df['total_profit'], color='Blue')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.xlabel('Company Profit per Month')
```

```python
plt.scatter(df['month_number'], df['toothpaste'])
plt.grid(linestyle = '--')
plt.show()
```
